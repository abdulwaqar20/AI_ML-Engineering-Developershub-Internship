# Task 4: General Health Query Chatbot

## Objective
Develop a prompt-engineered chatbot using Groq API that provides general health information.

## Model Used
- LLaMA 3.1 8B Instant (via Groq API)

## Features
- Friendly health assistant behavior
- Safety filtering to prevent harmful medical advice
- Interactive chatbot loop

## Disclaimer
This chatbot provides general information only and is not a substitute for professional medical consultation.

```python
import os
from getpass import getpass

os.environ["GROQ_API_KEY"] = getpass("Enter your GROQ API Key: ")
```

```python
from groq import Groq
import os

api_key = os.getenv("GROQ_API_KEY")

if not api_key:
    raise ValueError("API key not found!")

client = Groq(api_key=api_key)
```

```python
# Safety filter to block medical advice
def safety_filter(query):
    blocked_terms = [
        "diagnose", "prescribe", "dosage",
        "treatment", "cure", "antibiotic"
    ]
    return not any(term in query.lower() for term in blocked_terms)


def health_chatbot(user_query):
    prompt = f"""
You are a friendly and responsible medical assistant.
Provide general health information only.
Do NOT diagnose diseases or prescribe medicines.
If symptoms are serious, recommend consulting a doctor.

Question: {user_query}
Answer:
"""

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "system", "content": "You are a helpful medical assistant."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.5,
        max_tokens=200
    )

    return response.choices[0].message.content.strip()
```

```python
print("🩺 General Health Chatbot (Groq API - Colab)")
print("Type 'exit' to stop.\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Session ended.")
        break

    if safety_filter(user_input):
        reply = health_chatbot(user_input)
        print("Bot:", reply)
    else:
        print("⚠️ This question requires professional medical advice.")
```
